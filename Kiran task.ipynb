```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import math
import statsmodels.api as sm
import scipy.stats as stats


os.chdir('E:\\data science\\Mojotics\\Practice\\practice4 - task')

data = pd.read_csv('GermanCredit.csv')
```

```python
Y= data['Amount']
X = data.drop(['Amount'], axis=1)

data['Class'].replace('Bad',0,inplace=True)
data['Class'].replace('Good',1,inplace=True)

data['Class'].unique()
```

```python
selected_features_BE = ['Duration', 'InstallmentRatePercentage', 'Telephone', 'Class', 'CheckingAccountStatus.0.to.200', 'CheckingAccountStatus.none', 'CreditHistory.NoCredit.AllPaid', 'Purpose.UsedCar', 'Purpose.Radio.Television', 'Purpose.Other', 'SavingsAccountBonds.Unknown', 'Personal.Male.Single', 'OtherDebtorsGuarantors.CoApplicant', 'Property.Unknown', 'OtherInstallmentPlans.Bank', 'OtherInstallmentPlans.Stores', 'OtherInstallmentPlans.None', 'Job.Management.SelfEmp.HighlyQualified']
```

```python
matrix = pd.DataFrame()
for i in range(1,len(selected_features_BE)+1):
    matrix[i]= ""
    
print(matrix)
```

```python
for i in range(1,len(selected_features_BE)+1):
    for j in range(1,i+1):
        for k in range(j):
            matrix.loc[i,j] = selected_features_BE[k]
```

```python
print(matrix)
```

```python
X1 = sm.add_constant(X) # adding a constant
model = sm.OLS(Y, X1).fit()
predictions = model.predict(X1) 

print_model = model.summary()
print(print_model)
print(model.pvalues)
print(model.rsquared)
```

```python
r2_value=[]
adj_r2value =[]
fvalue =[]
aic =[]
bic=[]
durbinwatson_val =[]
MSE = []

for i in range(1,len(matrix)+1):
    
    c= matrix.loc[i,0:i]

    X = data[c]
    Y = data['Amount']
        
    X1 = sm.add_constant(X) # adding a constant
    model = sm.OLS(Y, X1).fit()
    predictions = model.predict(X1) 
    
    print_model = model.summary()
    r2_value.append(round(model.rsquared*100,2))
    
    adj_r2value.append(round(model.rsquared_adj*100,2))
    
    fvalue.append(round(model.fvalue,2))
    
    aic.append(round(model.aic,2))
    bic.append(round(model.bic,2))
    
    residuals = model.resid
    durbinwatson = sm.stats.stattools.durbin_watson(residuals, axis=0)
    durbinwatson_val.append(round(durbinwatson,2))
    
    MSE.append(model.mse_total)

    
matrix['r2_value'] = r2_value
matrix['adj_r2value'] = adj_r2value
matrix['fvalue'] = fvalue
matrix['aic'] = aic
matrix['bic'] = bic
matrix['durbinwatson_val'] = durbinwatson_val
matrix['MSE'] = MSE
```
